```python
from matplotlib import pyplot as plt
import random
```

```python
def getDate():
    x = range(2,26,2)
    y = [20+random.random()*10 for _ in range(12)]
    return x,y
```

### 调用方法plt.plot(x,y)，plt.show()绘图

```python
x,y = getDate()
plt.plot(x,y)
plt.show()
```

### 设置图片大小

```python
x,y = getDate()
fig = plt.figure(figsize=(16,9),dpi=100)
plt.plot(x,y)
plt.savefig("./tmp/1.png") # 可以使用".svg"矢量图格式保存
```

**savefig(file_path)用于保存图片**

### 设置x,y轴刻度

```python
x,y = getDate()
plt.figure(figsize=(16,9),dpi=100)
plt.xticks([i/2 for i in range(4,49)][::3]) #tick是记号的意思
plt.yticks(range(30))
plt.plot(x,y)
plt.show()
```

### 中文支持
Windows下字体在C:\Windows\Fonts,或运行->fonts；在Linux/mac下通过`fc-list`查看支持的字体，用`fc-list :lang=zh`查看支持的中文字体（冒号前有空格）

```python
from matplotlib import font_manager

my_font = font_manager.FontProperties(fname='C:\Windows\Fonts\simsun.ttc')

x = range(120)
y = [random.randint(20,35) for _ in range(120)]
plt.figure(figsize=(16,9),dpi=100)
plt.plot(x,y)
_xticks = ["10点{}分".format(i) for i in range(60)]+["11点{}分".format(i) for i in range(60)] #这时坐标轴上的数字和x数据不同，需要指定替换位置和替换的值
plt.xticks(list(x)[::3],_xticks[::3],rotation=60,fontproperties=my_font) # rotation=90：x坐标轴刻度旋转90度
plt.show()
```

### 添加描述信息

```python
def addDescribe():
    plt.xlabel("时间",fontproperties=my_font)
    plt.ylabel("温度 单位（℃）",fontproperties=my_font)
    plt.title("一天气温变化图",fontproperties=my_font)
    
    plt.xticks(range(0,25))
def play():
    x,y = getDate()
    plt.plot(x,y)
    addDescribe()
play()
plt.show()
```

### 绘制网格

```python
play()
plt.grid(alpha=0.1)# alpha：透明度
plt.show()
```

### 绘制多条曲线

```python
x,y = getDate()
plt.plot(x,y)

x1,y1 = getDate()
plt.plot(x1,y1)

plt.xlabel("时间",fontproperties=my_font)
plt.ylabel("温度 单位（℃）",fontproperties=my_font)
plt.title("一天气温变化图",fontproperties=my_font)
plt.xticks(range(0,25))
plt.show()
```

### 添加图例

```python
x,y = getDate()
plt.plot(x,y,label="合肥")

x1,y1 = getDate()
plt.plot(x,y,label="六安")

addDescribe()

plt.legend(prop=my_font)
plt.show()
```
