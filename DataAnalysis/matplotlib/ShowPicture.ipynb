```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open("1.jpg")
plt.figure("Image") # 定义窗口名称
plt.imshow(img)
plt.axis('on') # 关掉坐标轴为off
plt.title('image')
plt.show()
```

```python
# 显示多张图片
plt.figure()
plt.subplot(2,3,1)
plt.imshow(img)
plt.title("1")
plt.subplot(2,3,2)
plt.imshow(img)
plt.title("2")
plt.subplot(2,3,3)
plt.imshow(img)
plt.title("3")
plt.subplot(2,3,4)
plt.imshow(img)
plt.title("4")
plt.subplot(2,3,5)
plt.imshow(img)
plt.title("5")s
plt.subplot(2,3,6)
plt.imshow(img)
plt.title("6")
plt.show()
```
